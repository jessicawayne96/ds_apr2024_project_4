```python
# Import the required modules
import pandas as pd
pd.set_option('display.max_columns', None)
import numpy as np

# visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# The paths to the CSV files
filepath1 = "Resources/heroes_information.csv"
filepath2 = "Resources/super_hero_powers.csv"

#reading the csv into pandas
hero_info = pd.read_csv(filepath1)
hero_powers = pd.read_csv(filepath2)

print(hero_info.shape)
print(hero_powers.shape)
```

## Hero Information cleaning 

```python
hero_info.head(15)
```

```python
hero_info.info()
```

```python
#handeling the nulls
hero_info.isnull().sum()
```

```python
hero_info.columns
```

```python
#nulls are a little sus, looking at the .head there look to be nulls in the eyes/race/skin color/ hair color
# Replace hyphens with NaN (thank you expert)
#trying eye color to make sure this works
hero_info['Eye color'].replace('-', np.nan, inplace=True)
hero_info.head(15)
```

```python
# Replace hyphens with NaN for race now 
hero_info['Race'].replace('-', np.nan, inplace=True)
hero_info.head(15)
```

```python
# Replace hyphens with NaN for skin color now 
hero_info['Skin color'].replace('-', np.nan, inplace=True)
hero_info.head(15)
```

```python
# Replace hyphens with NaN for skin color now 
hero_info['Hair color'].replace('-', np.nan, inplace=True)
hero_info.head(15)
```

```python
#making a copy of the data frame so i can drop all of the rows with NaN 
hero_info_clean1 = hero_info.copy()
```

```python
#dropping all the NaN values in the copy of the data frame
hero_info_clean1 = hero_info_clean1.dropna()
hero_info_clean1.head(15)
```

```python
hero_info_clean1.info()
```

```python
#dropping the null rows got rid of 90% of the data.....
#making a second copy and dropping the columns that I changed the hyphens to nulls
hero_info_clean2 = hero_info.copy()
```

```python
hero_info_clean2 = hero_info_clean2.drop(['Race','Eye color','Hair color','Skin color'], axis =1)
hero_info_clean2.info()
```

```python
# NOW dropping the few nulls that are left
hero_info_clean2 = hero_info_clean2.dropna()
hero_info_clean2.info()
```

```python
# saving the new cleaned csv for tableau
hero_info_clean2.to_csv('clean_hero_info.csv', index=False)
```
